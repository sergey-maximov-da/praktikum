Прочитаем таблицу по ссылке

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
tables = pd.read_html(
  'https://worldtable.info/gosudarstvo/tablica-rozhdaemosti-po-godam-rossija.html')
```

Запилим list в ДатаФрейм, удалим пробелы в поле counts, а также переведем в формат целого числа


```python
df =tables[0]
df = df.drop(labels = [0],axis =0)
df.columns = ['year','counts']
df['counts'] = df['counts'].replace(r'\s+','',regex=True).astype('int')
df
```

Класс! Построим график рождаемости по годам в России



```python
df.plot(x='year',y='counts',figsize=(18,6),title='Таблица рождаемости в России по годам',xlabel='Год',ylabel='Количество родившихся',fontsize=14)
plt.show()
```

Ну что же... Пиковые значения рождаемости приходятся на начало наблюдений - 1959 год. Скорее всего это связано с ростом и востановлением экономики после войны. Снижение рождаемости к 1969 году может быть связано с предыдущим бумом рождаемости - взяли паузу. С 1970 по 1989 год наблюдается уверенный рост годовой рождаемости - стабильность в стране дает уверенность будущим родителям. В связи с надвигающимся развалом СССР и продолжительным упадком экономики с 1989 по 1999 года наблюдаем абсолютный минимум рождаемости. С приходом к власти того, ког нельзя называть - экономика России начинает расти и у людей появляется уверенность в завтрашнем дне ( от слова дно), увереность продолжается вплоть до 2014 года, далеее идет спад очень похожий на 1990 год...
