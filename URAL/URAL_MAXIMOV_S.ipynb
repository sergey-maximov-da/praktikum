#### Задание: вывести количество действующих предложений на каждый день 2018 года.

```python
sql = ''' 
              SELECT DATE(OFFER_START_DATE) as DATE,
                     offer_id,
                     COUNT (offer_id) AS VALLID_OFFER
              FROM offer
              WHERE OFFER_START_DATE BETWEEN '2017-01-01 00:00:00' AND '2018-12-30 00:00:00'
              AND OFFER_EXPIRATION_DATE <= '2018-12-30 00:00:00' 
              GROUP BY DATE(OFFER_START_DATE)
          
      '''
select(sql)
```

```python
sql = ''' 
          SELECT DATE_START, 
                 SUM(OFFERS) over (order by DATE_START rows between unbounded preceding and current row) AS COUNT_OFFERS
          FROM
             (SELECT DATE(OFFER_START_DATE) as DATE_START,
                     COUNT(offer_id) as OFFERS
                  
              FROM offer
              WHERE OFFER_START_DATE BETWEEN '2017-01-01 00:00:00' AND '2018-12-31 00:00:00'
                    AND OFFER_EXPIRATION_DATE <= '2018-12-31 00:00:00 ' 
              GROUP BY DATE(OFFER_START_DATE))
            
      '''
select(sql)
```

### Решение 1

```python
sql = '''SELECT client_id,
                card_id,
                open_date
         FROM CARD
         WHERE OPEN_DATE < '2018-09-01 00:00:00' AND CLOSE_DATE is null AND CARD_TYPE = '1'
         ORDER BY (OPEN_DATE) DESC
    '''
select(sql)
```

### Решение 2

```python
sql = '''
          SELECT CLIENT_ID, CARD_ID
          FROM 
          
           ( SELECT CLIENT_ID,
                    CARD_ID,
                    OPEN_DATE,
                    MAX(OPEN_DATE),
                    CLOSE_DATE,
                    CARD_TYPE,
                    row_number() over (partition by CLIENT_ID order by OPEN_DATE) AS rn
             FROM CARD
             WHERE OPEN_DATE < '2018-09-01 00:00:00' AND CLOSE_DATE is null AND CARD_TYPE = '1')
             
           WHERE rn=1
           ORDER BY OPEN_DATE
      '''
select(sql)
```

### Кроме того, что мой импровизированный датасет не позволил мне найти клиентов с 1000 баллами, таких просто нет, поэтому я остановился на 500 баллах, и 1000 клиентов с 500 баллами тоже не набралось, но тем не менее этот запрос отработает, если поменять where NAR >= *** )

```python
sql = ''' SELECT CLIENT,
                 DATE,
                 NAR
          FROM
            (SELECT CLIENT,
                    DATE,
                    BONUS,
                    NAR,
                    ROW_NUMBER () OVER (PARTITION BY CLIENT order by DATE ) AS ROW
             FROM
                 (SELECT CLIENT_ID AS CLIENT,
                         BONUS_DATE AS DATE,
                         BONUS_CNT AS BONUS,
                         SUM(BONUS_CNT) over (PARTITION BY CLIENT_ID 
                                          order by BONUS_DATE rows between unbounded preceding and current row) AS NAR
                  FROM bonus JOIN mcc_categories ON bonus.MCC = mcc_categories.MCC
                  WHERE MCC_CATEGORY IN ('Авиабилеты',
                                        'Отели')
                  ORDER BY CLIENT_ID)
                  WHERE NAR >= 500
                  ORDER BY DATE)
           WHERE ROW = 1
           LIMIT 1000
      '''
select(sql)
```
